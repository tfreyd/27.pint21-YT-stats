# retrieve views counts per videos:

```python
def GetVideoId(YoutubeChannelId):
    list_ids_videos=[]
    
    #setting the parameters
    api_key = YT_key  # Please set your API key

    from googleapiclient.discovery import build
  
    api_service_name = "youtube"
    api_version = "v3"
    youtube = build(api_service_name, api_version, developerKey=api_key)
    request = youtube.search().list(
        part="snippet",
        publishedAfter='2021-05-16T00:00:00Z', 
        channelId=YoutubeChannelId,
        type='video',
        maxResults=100,
    )
    response = request.execute()
    print(response)

    #extract videoID
    for list_videos in response['items']:
        id_video=list_videos['id']['videoId']
        print(id_video)
        list_ids_videos.append(id_video)
    #print((response))

    if 'nextPageToken' in response.keys():
        nextPageToken=response['nextPageToken']

        request2 = youtube.search().list(
        part="snippet",
        publishedAfter='2021-05-16T00:00:00Z', 
        channelId=YoutubeChannelId,
        maxResults=100,
        type='video',
        pageToken=nextPageToken
        )

        response2 = request2.execute()

        for list_videos in response2['items']:
            id_video=list_videos['id']['videoId']
            print(id_video)
            list_ids_videos.append(id_video)
    else:
        pass

    return list_ids_videos
```

```python
def getVideoViews(list_video_id):
    total_views=0
    for video_id in list_video_id:

        request = youtube.videos().list(
            part="snippet,statistics",
            id=video_id,
        )
        response = request.execute()
        #extract and cumulate the view
        views=response['items'][0]['statistics']['viewCount']
        total_views = total_views + int(views)

    return total_views
```
